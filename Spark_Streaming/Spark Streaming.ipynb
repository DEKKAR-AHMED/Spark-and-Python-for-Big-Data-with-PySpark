```python
from pyspark import SparkContext
from pyspark.streaming import StreamingContext
```

```python
sc = SparkContext("local[2]","NetworkWordCount") # local 2 indicates the number of thread
```

```python
scc = StreamingContext(sc,1) # the 1 indicates that the interval between threads is one secondes 
```

```python
# to connect to jupyter notebook and our terminal  as source of streaming data 
lines = scc.socketTextStream("localhost",9999)
```

```python
# turn the line that we grab to a list of words 
words = lines.flatMap(lambda line: line.split(" "))
```

```python
# make pairs of word 
pairs = words.map(lambda word: (word,1))
```

```python
# count the number of words in the sentence
word_count = pairs.reduceByKey(lambda num1, num2 : num1 + num2 )
```

```python
word_count.pprint()
```
